### Import the library for word cloud and the STOPWORDS to eliminate words usually made up of articles, pronouns, prepositions, adverbs, and even some verbs.

```python
from wordcloud import WordCloud, STOPWORDS
```

#### Load the book (SPANISH VERSION)

```python
quijote = open('don_quijote.txt', 'rb').read().decode('utf-8')
```

#### use the custom stopwords spanish list and add it to a new list

```python
stop_words = []
with open("stop_words_spanish.txt", "r") as f:
    word = f.read()
    stop_words = word.splitlines()
```

```python
stop_words[:10]
```

#### Add new stopwords

```python
stop_words.extend(['él', 'qué', 'así', 'allí', 'te', 'cómo', 'más',
                   'había', 'día', 'aquí', 'tu', 'está', 'os', 'quién',
                   'mis', 'tú', 'sé', 'aun', 'mí', 'vos', 'ma', 'has',
                  'algún'])
```

#### Run the WordCloud function and customize max words, color and stopwords

```python
don_quijote_wc = WordCloud(
        background_color='white',
        max_words=200,
        stopwords=stop_words
)

don_quijote_wc.generate(quijote)
```

#### Plot the resulted words

```python
plt.imshow(don_quijote_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

#### Open the silhouette image of Don Quixote

```python
don_quixote_mask = np.array(Image.open("silhouette.jpg"))
```

#### Give size and show the image

```python
fig = plt.figure(figsize=(14, 18))
plt.imshow(don_quixote_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.show()
```

### Create the Wordcloud in the silhouette image

```python
quixote_wc = WordCloud(
        background_color='white',
        max_words=200,
        mask=don_quixote_mask,
        stopwords=stop_words
)

quixote_wc.generate(quijote)
fig = plt.figure(figsize=(14, 18))
plt.imshow(quixote_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
